```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("orders_data.csv")
df
# successfully loads the dataframe
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.dropna()
# dropping any null values
```

```python
# checking for duplicates and dropping them
df.drop_duplicates()
```

```python
df.drop(columns={"City"},inplace=True)
# did this because i do not need in depth location of the sales
```

```python
df.head(1)
```

```python
df['List Price'] = df['List Price'] * ((100 - df['Discount Percent'])/100)
```

```python
df.head(5)
```

```python
df['Profit'] = df['List Price'] - df['cost price']
```

```python
df.head(5)
```

```python
df['Total Profit'] = df['Profit'] * df['Quantity']
```

```python
df.head(5)
```

```python
df.drop(columns={'Profit','Discount Percent'},inplace=True)
```
